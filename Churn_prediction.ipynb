```python
def tenure(tenure_in_months):
  if tenure_in_months<12:
    return 'New'
  elif tenure_in_months<36:
    return 'Regular'
  else:
    return 'Churned'

df['tenure_segment']=df['tenure_in_months'].apply(tenure)
df['tenure_segment'].value_counts()
```
